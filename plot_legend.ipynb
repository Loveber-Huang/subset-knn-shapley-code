```python
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

GLOBAL_METHOD_COLORS = {
    "Inc-CkNN-E": "#ff7f0e",
    "Random": "#9467bd",
    "Inc-CkNN-Apx": "#2ca02c",
    "InRun-DataShapley": "#7f7f7f",
    "Inc-CkNN-ANN": "#d62728",
    "BetaShapley": "#bcbd22",
    "KNN-Shapley": "#1f77b4",
    "Influence Function": "#8c564b",
    "TMC-DataShapley": "#e377c2",
    "CS-Shapley": "#17becf",
    "LAVA": "#393b79",
}

GLOBAL_METHOD_MARKERS = {
    "Inc-CkNN-E": "o",
    "Random": "s",
    "Inc-CkNN-Apx": "v",
    "InRun-DataShapley": "*",
    "Inc-CkNN-ANN": "^",
    "BetaShapley": "<",
    "KNN-Shapley": "D",
    "Influence Function": "P",
    "TMC-DataShapley": "X",
    "CS-Shapley": ">",
    "LAVA": "h",
}


handles = [
    Line2D(
        [0], [0],
        color=color,
        marker=GLOBAL_METHOD_MARKERS.get(method),
        linewidth=2,
        markersize=8,
        label=method
    )
    for method, color in GLOBAL_METHOD_COLORS.items()
]

fig, ax = plt.subplots(figsize=(12, 2))
ax.axis("off")

ax.legend(
    handles=handles,
    loc="center",
    ncol=6,          # 两行
    frameon=False,
    fontsize=12,
    handlelength=2.2,
    columnspacing=1.6,
    handletextpad=0.5,
)

plt.tight_layout()
plt.savefig("method_legend.pdf", format="pdf", bbox_inches="tight")
plt.show()
```
