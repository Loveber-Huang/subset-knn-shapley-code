```python
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D
import os

GLOBAL_METHOD_COLORS = {
    "Inc-CkNN-E": "#ff7f0e",
    "Inc-CkNN-Apx": "#2ca02c",
    "Inc-CkNN-ANN": "#d62728",
    "Random": "#9467bd",
    "KNN-Shapley": "#1f77b4",
}

GLOBAL_METHOD_MARKERS = {
    "Inc-CkNN-E": "o",
    "Inc-CkNN-Apx": "v",
    "Inc-CkNN-ANN": "^",
    "Random": "s",
    "KNN-Shapley": "D",
}


handles = [
    Line2D(
        [0], [0],
        color=color,
        marker=GLOBAL_METHOD_MARKERS.get(method),
        linewidth=2,
        markersize=8,
        label=method
    )
    for method, color in GLOBAL_METHOD_COLORS.items()
]

fig, ax = plt.subplots(figsize=(6, 2))
ax.axis("off")

ax.legend(
    handles=handles,
    loc="center",
    ncol=6,          # 两行
    frameon=False,
    fontsize=12,
    handlelength=2.2,
    columnspacing=1.6,
    handletextpad=0.5,
)

os.makedirs("pics", exist_ok=True)

plt.tight_layout()
plt.savefig("pics/method_legend.pdf", format="pdf", bbox_inches="tight")
plt.show()
```
